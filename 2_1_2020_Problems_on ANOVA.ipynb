```python
# Ex 13-7
import pandas as pd
import numpy as np
import math
from scipy import stats
import scipy
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel('compressive strength.xlsx')
```

```python
data
```

```python
data_new=pd.melt(data.reset_index(),id_vars=['index'],value_vars=['MT1','MT2','MT3','MT4'])
data_new.columns=['index','Treatments','value']
data_new
```

```python
model=ols('value~C(Treatments)',data=data_new).fit()
anova_table=sm.stats.anova_lm(model,type=1)
anova_table
```

```python
from scipy import stats
p_value=1-stats.f.cdf(12.728107,3,12)
p_value
```

```python
scipy.stats.f.ppf(1-0.05,dfn=3,dfd=12)
```

```python
#Since p_value is than alpha and F_computed is greater than Falpha we reject H0.
```
