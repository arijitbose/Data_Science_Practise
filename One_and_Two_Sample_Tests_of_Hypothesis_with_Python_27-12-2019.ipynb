```python
# A statistical hypothesis is am assertion or conjecture conerning one or more populations.
# Often the problem confronting the scientist or engineer is not so much the estimation of a population parameter, but rather 
# formation of a data based decision procedure that can produce a conclusion about the the population concerned.
```

#First we concentrate on errors in Hypotheses testing
Example: We are interested in burning rate of a solid propellant used to power air crew escape systems.
Burning rate is a random variable that can be described by a probability distribution.
Suppose our interest focus on mean burning rate.
H0: mu =  50 centimeters per second
H1: mu != 50 centimeters per second




#sigma = 2.5
#n=number of samples = 10

```python
import numpy as np
from scipy.stats import norm
```

```python
def z_value(x,mu,sigma):
    z=(x-mu)/sigma
    if(z<0):
        alpha=norm.cdf(z)
    else:
        alpha=1-norm.cdf(z)
    return alpha    
```

```python
sigma= 2.5/np.sqrt(10)
sigma
```
