# Hypothesis testing can be used to determine whether a statement about the
# value of a population parameter should or should not be rejected.

# The null hypothesis denoted by H0,is a tentative assumption about a population parameter.

# The alternative hypothesis denoted by Ha, is the opposite of what is stated in # the null hypothesis.
# The hypothesis testing procedure uses data from a sample to test two competing statements indicated by H0 and Ha.


# Developing NULL and Alternative Hypotheses
# Alternative Hypothesis as a research Hypothesis
# Example: A new manufacturing method is belived to better than the current method.
# Alternative Hypothesis: The new manufacturing method is better.
# Null Hypothesis: The new method is no better than the old method.


# H0 : mu>=mu0
# Ha :  mu<mu0
# This form of Hypotheses is One-tailed(lower-tail)


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
x_axis = np.arange(47, 54, 0.00001)
# Mean = 0, SD = 2.
# x coordinates for the lines
xcoords = [48.5, 50, 52.5]
# colors for the lines
colors = ['r','k','y']

for xc,c in zip(xcoords,colors):
    plt.axvline(x=xc, label='line at x = {}'.format(xc), c=c)
    plt.axvspan(52.5, 54, alpha=0.5, color='red')
    plt.axvspan(46, 48.5, alpha=0.5, color='red')
    plt.axvspan(48.5,52.5,alpha=0.5, color='green')
    
plt.plot(x_axis, norm.pdf(x_axis,50,0.7905694150420948),color='yellow')



plt.show()




```

# p-Value Approach

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
x_axis = np.arange(-5,5 , 0.00001)
# Mean = 0, SD = 2.
# x coordinates for the lines
xcoords = [-1.46, -1.28, 0]
# colors for the lines
colors = ['r','k','y']

for xc,c in zip(xcoords,colors):
    plt.axvline(x=xc, label='line at x = {}'.format(xc), c=c)
    #plt.axvspan(52.5, 54, alpha=0.5, color='red')
    plt.axvspan(-1.28,6 ,alpha=0.5, color='green')
    plt.axvspan(-5,-1.46,alpha=0.5, color='red')
    
plt.plot(x_axis, norm.pdf(x_axis,0,0.7905694150420948),color='yellow')



plt.show()
```

```python
# P-value Approach, lower-Tailed Test About A Population Mean : sigma is known
# we have xbar,mu0,sigma and n and hence Z and alpha values
# green color is acceptance region
# Red color: rejection region
from scipy.stats import norm
p_value = norm.cdf(-1.46)
p_value
```

```python
# Upper tail test About a population mean: sigma known, p value approach
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
x_axis = np.arange(-5,5 , 0.00001)
# Mean = 0, SD = 2.
# x coordinates for the lines
xcoords = [0, 1.75, 2.29]
# colors for the lines
colors = ['r','k','y']

for xc,c in zip(xcoords,colors):
    plt.axvline(x=xc, label='line at x = {}'.format(xc), c=c)
    #plt.axvspan(52.5, 54, alpha=0.5, color='red')
    plt.axvspan(2.29,6 ,alpha=0.5, color='red')
    plt.axvspan(-5,1.75,alpha=0.5, color='green')
    
plt.plot(x_axis, norm.pdf(x_axis,0,0.7905694150420948),color='yellow')



plt.show()
```

```python
# Critical Value approach to One tailed Hypothesis testing
# Corresponding Z value is retreived in python as follows
# alpha = 0.1
z_critical=norm.ppf(0.1)
z_critical
```

```python
# Page 380 Ronald book
#Ex3
#A random sample of 100 recorded deaths in the United States during the past year showed an average life span of 71.8 years.
#Assuming a population standard deviation of 8.9 years,does this seem to indicate that mean life span today is greater than
#70 years? Use alpha, the level of significance as 0.05.
#Solution
#1.H0: mu=70 years
#2.H1: mu>70 years
import numpy as np
from scipy.stats import norm
p_value = norm.cdf(-1.46)
alpha=0.05
z_critical=abs(norm.ppf(alpha))
z_computed=((71.8-70)/8.9)*np.sqrt(100)# calculated from given data using formula ((xbar-mu)/sigma)*(n)**(0.5)
p_value = 1-norm.cdf(z_computed)# using one minus cdf function since it is upper tail test, p_value is the area inder the curve
                                # towards the right of z_computed  
def upper_tail_test(p_value,alpha):
    if(p_value<alpha):
        return "accept H1"
    else:
        return "accept H0"
upper_tail_test(p_value,alpha)    
    
    


```
